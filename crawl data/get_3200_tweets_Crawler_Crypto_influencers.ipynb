```python
#reddit

d1=pd.read_csv('reddit_data.csv', encoding='UTF-8',dtype='str')
d1=d1.rename(columns={'Unnamed: 0'	:'index'})
main_df=d1

coins=pd.read_csv('Top_100_coinMarketCap.csv')
coins['with$']=coins['symbol'].apply(lambda x:'$'+x+' ')
coins['with#']=coins['symbol'].apply(lambda x:'#'+x+' ')
coins['symbol']=coins['symbol'].apply(lambda x:' '+x.upper()+' ')
coins.iloc[54]['symbol']='$one '

main_df['concat_fields']='A'
for i ,notn in enumerate((main_df['selftext'].notna())):
  try:
    if notn:
      main_df['concat_fields'][i]=(main_df.iloc[i,1]+ ' selftext: ' + main_df.iloc[i,6]  )
    else:
      main_df['concat_fields'][i]=(main_df.iloc[i,1] )
  except:
    print(i)
# data['concat_fields']=np.array(temp)
main_df
```

```python
punctuations='''@!"%'()*+,-./:;<=>?[\]^_`{|}~'''
main_df['clean_tweet'] = main_df['concat_fields'].apply(lambda x: unidecode.unidecode(x).translate(str.maketrans('', '', punctuations)))
```

```python
main_df['coins'] = main_df['clean_tweet'].apply(lambda x :  [coins['coin_name'][i] for i in range(len(coins)) if (coins['coin_name'][i]+' ' in x.lower()) or (coins['symbol'][i] in x) or (coins['with$'][i] in x.lower()) or (coins['with#'][i] in x.lower())] )
print('The number of tweets that has coin names is :',len(main_df))
```

```python
for i in range(len(main_df['coins'])):
  if ' btc ' in main_df.iloc[i]['clean_tweet'] and 'bitcoin' not in main_df.iloc[i]['coins'] :
    main_df.iloc[i]['coins'].append('bitcoin')
for i in range(len(main_df['coins'])):
  if 'crypto' in main_df.iloc[i]['clean_tweet']  :
    main_df.iloc[i]['coins'].append('crypto')
```

```python
clean_main_df2=main_df[main_df['coins'].apply(lambda x:len(x)!=0)]
clean_main_df2['coins_name'] = clean_main_df2['clean_tweet'].apply(lambda x :  [coins['coin_name'][i] for i in range(len(coins)) if (' '+coins['coin_name'][i]+' ' in x.lower()) or ('$'+coins['coin_name'][i]+' 'in x.lower()) or ('#'+coins['coin_name'][i]+' ' in x.lower())] )
clean_main_df2['coins_symbol'] = clean_main_df2['clean_tweet'].apply(lambda x :  [coins['coin_name'][i] for i in range(len(coins)) if (coins['symbol'][i] in x)                    or (coins['with$'][i] in x.lower())            or (coins['with#'][i] in x.lower())] )

print('The number of tweets that has coin names is :',len(clean_main_df2))
```

```python
unrelated_keys=['airdrop','giveaway']
clean_main_df2['unrelated_keys'] = clean_main_df2['clean_tweet'].apply(lambda x :  [unrelated_keys[i] for i in range(len(unrelated_keys)) if (unrelated_keys[i] in x.lower())] )
print('The number of tweets that has coin names is :',len(clean_main_df2))
```
