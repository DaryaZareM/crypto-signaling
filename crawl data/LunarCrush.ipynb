```python
from lunarcrush import LunarCrush
import pandas as pd
from pprint import pprint
import json
```

```python
coins=pd.read_csv('top_MCcoin_15.csv')
```

```python
coins
```

```python
lc = LunarCrush()
for csymbol in coins['symbol']:
  twitter_influencers=lc.get_influencers(symbol=[csymbol],days=60)
  with open(f'lunarcrush_influencers/data60_{csymbol}.json', 'w') as fp:
    json.dump(twitter_influencers, fp)
```

```python
import json
mdf=pd.DataFrame()
for csymbol in coins['symbol']:
  x = json.load(open(f'lunarcrush_influencers/data60_{csymbol}.json'))

  try:
    df=pd.DataFrame(x['data'])
    df['coin_name']=[csymbol]*(len(df))
    mdf=pd.concat([mdf,df],ignore_index=True)
  except:
    print(csymbol)
mdf
```
